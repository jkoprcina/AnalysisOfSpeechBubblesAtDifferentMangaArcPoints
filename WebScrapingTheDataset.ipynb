```python
# Imports for Web Scraping
import os
import requests
from bs4 import BeautifulSoup as soup
from urllib.request import Request, urlopen
```

```python
# manga url's
NARUTO_URL = "https://naruto-shippudenmanga.com/manga/naruto-chapter-"
BOKU_NO_HERO_URL = "https://w21.readheroacademia.com/manga/boku-no-hero-academia-chapter-"
ONE_PIECE_URL = "https://one-piece-online-manga.com/manga/one-piece-chapter-"
# needed json object to web scrape
usr_agent = {
    'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.11 (KHTML, like Gecko) Chrome/23.0.1271.64 Safari/537.11',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
    'Accept-Charset': 'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
    'Accept-Encoding': 'none',
    'Accept-Language': 'en-US,en;q=0.8',
    'Connection': 'keep-alive',
} 

ROUTE = "manga"
if not os.path.exists(ROUTE):
    os.mkdir(ROUTE)
```

```python
# Download necessary naruto chapters
MANGA_ROUTE = ROUTE + "/naruto"
if not os.path.exists(MANGA_ROUTE):
    os.mkdir(MANGA_ROUTE)

for i in range(1,239):
    print("Naruto chapter " + str(i) + " downloading...")
    if(i == 1):
        request = Request("https://naruto-shippudenmanga.com/manga/naruto-shippuden-chapter-1/", headers=usr_agent)
    else:
        request = Request(NARUTO_URL + str(i) + "/", headers=usr_agent)
        
    webpage = urlopen(request).read()
    page_soup = soup(webpage, "html.parser")
    results = page_soup.findAll('img')
    image_links = [img.get("src") for img in results]
    
    CHAPTER_ROUTE = MANGA_ROUTE + "/chapter" + str(i)
    if not os.path.exists(CHAPTER_ROUTE):
        os.mkdir(CHAPTER_ROUTE)
    
    for j, image_link in enumerate(image_links):
        response = requests.get(image_link)
        if(j < 9):
            imagename = CHAPTER_ROUTE + "/page0" + str(j+1) + '.jpg'
        else:
            imagename = CHAPTER_ROUTE + "/page" + str(j+1) + '.jpg'
        with open(imagename, 'wb') as file:
            file.write(response.content)
            
print("Naruto chapters have downloaded")
```

```python
# Download necessary boku no hero chapters
MANGA_ROUTE = ROUTE + "/boku-no-hero"
if not os.path.exists(MANGA_ROUTE):
    os.mkdir(MANGA_ROUTE)
    
for i in range(1,184):
    print("Boku no hero chapter " + str(i) + " downloading...")
    if(i == 1):
        request = Request("https://w21.readheroacademia.com/manga/boku-no-hero-academia-chapter-one/", headers=usr_agent)
    else:
        request = Request(BOKU_NO_HERO_URL + str(i) + "/", headers=usr_agent)
    webpage = urlopen(request).read()
    page_soup = soup(webpage, "html.parser")
    results = page_soup.findAll('img') 
    image_links = [img.get("src") for img in results]
    
    CHAPTER_ROUTE = MANGA_ROUTE + "/chapter" + str(i)
    if not os.path.exists(CHAPTER_ROUTE):
        os.mkdir(CHAPTER_ROUTE)
    
    for j, image_link in enumerate(image_links):
        response = requests.get(image_link)    
        if(j < 10):
            imagename = CHAPTER_ROUTE + "/page0" + str(j+1) + '.jpg'
        else:
            imagename = CHAPTER_ROUTE + "/page" + str(j+1) + '.jpg'
        with open(imagename, 'wb') as file:
            file.write(response.content)
            
print("All boku no hero chapters have downloaded")
```

```python
# Download necessary one piece chapters
MANGA_ROUTE = ROUTE + "/one-piece"
if not os.path.exists(MANGA_ROUTE):
    os.mkdir(MANGA_ROUTE)
    
for i in range(1,218):
    print("One Piece chapter " + str(i) + " downloading...")
    request = Request(ONE_PIECE_URL + str(i) + "/", headers=usr_agent)
    webpage = urlopen(request).read()
    page_soup = soup(webpage, "html.parser")
    results = page_soup.findAll('img') 
    image_links = [img.get("src") for img in results]
    image_links = [img for img in image_links if img[-4:] == ".jpg"]
    
    CHAPTER_ROUTE = MANGA_ROUTE + "/chapter" + str(i)
    if not os.path.exists(CHAPTER_ROUTE):
        os.mkdir(CHAPTER_ROUTE)
    
    for j, image_link in enumerate(image_links):
        response = requests.get(image_link)    
        if(j < 10):
            imagename = CHAPTER_ROUTE + "/page0" + str(j+1) + '.jpg'
        else:
            imagename = CHAPTER_ROUTE + "/page" + str(j+1) + '.jpg'
        with open(imagename, 'wb') as file:
            file.write(response.content)
            
print("All One Piece chapters have downloaded")
```
